```python
def load_Math23K_data_withgroup(filename,group_file):  # load the json data to list(dict()) for MATH 23K
    print("Reading lines...")

    with open(group_file,'r',encoding='utf-8') as f1:
        group = json.load(f1)

    #if USE_dependency==True:
    with open("./Math23K_edge.json","r",encoding='utf-8') as f2:
        edge=json.load(f2)

    f = open(filename,'r', encoding='utf-8')
    js = ""
    data = []
    g_id=0
    e_id=0
    cout_no_g=0
    count_no_e=0
    for i, s in enumerate(f):
        js += s
        i += 1
        if i % 7 == 0:  # every 7 line is a json
            data_d = json.loads(js)

            question, equation, answer = data_d['segmented_text'].strip(), data_d['equation'], data_d['ans']


            equation = equation.replace('"千米/小时"', '')
            if equation[:2] == 'x=':
                equation = equation[2:]
            js = ""

            id_=data_d['id']
            g=group[g_id] #
            ed_=edge[e_id] #
            group_this=[]
            list_dependency=[]
            if id_==g["id"]:
                group_this=g["group_num"]
                g_id+=1
            else:
                print("no group")
                cout_no_g+=1
            if id_==ed_["id"]:
                list_dependency=ed_["list_dependency"]
                e_id+=1
            else:
                print("no dependency")
                count_no_e+=1

            data.append((question, equation, answer,group_this,list_dependency))
    print(cout_no_g)
    print(count_no_e)
    return data
```

```python
data = load_Math23K_data_withgroup('./Math_23K.json',"./Math_23K_my_processed.json")
```
